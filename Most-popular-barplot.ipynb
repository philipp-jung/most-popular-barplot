```python
from collections import Counter
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
survey = pd.read_csv('survey_results_public.csv', sep=',', na_values=np.nan)
```

```python
language_counter = Counter()
for row in survey.loc[:, 'LanguageWorkedWith'].values:
    row = str(row) # just count nans
    language_counter.update(row.split(';'))
```

```python
lang, popularity = zip(*language_counter.most_common(10))
```

```python
lang = list(lang)
popularity = list(popularity)
```

```python
lang.reverse()
popularity.reverse()
```

```python
plt.style.use('fivethirtyeight')
plt.barh(lang, popularity)
plt.title("Most used Programming Languages 2019")
plt.xlabel("Number of Users")
#plt.show()
plt.savefig('./popular-languages-2019.png', dpi=300, bbox_inches='tight')
```
